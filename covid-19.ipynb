```python
country_name = 'United States'

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
fig.set_figheight(8)
fig.set_figwidth(20)
fig.suptitle('Covid-19 Vaccination Data Analysis', fontsize=30)
fig.tight_layout(h_pad=7, w_pad=10)

ax1.set_title('Vaccination Rate', fontsize=22)
sns.lineplot(x=vaccine_data['date'], y=vaccine_data[vaccine_data['location']==country_name]['new_people_vaccinated_smoothed'], ax=ax1, color="purple")
ax1.set_xlabel('Date', fontsize=18)
ax1.set_ylabel('New Vaccinations', fontsize=18)

ax2.set_title('Vaccinations', fontsize=22)
sns.lineplot(x=vaccine_data['date'], y=vaccine_data[vaccine_data['location']==country_name]['people_vaccinated'], ax=ax2, color="blue")
ax2_1 = ax2.twinx()
sns.lineplot(x=vaccine_data['date'], y=vaccine_data[vaccine_data['location']==country_name]['people_fully_vaccinated'], ax=ax2_1, color="darkblue")
ax2.set_xlabel('Date', fontsize=18)
ax2.set_ylabel('People with one dose', fontsize=18)
ax2_1.set_ylabel('People Fully Vaccinated', fontsize=18)
ax2.legend(['People with one dose', 'People Fully Vaccinated'])

plt.subplots_adjust(top=0.90)

plt.show()
```
